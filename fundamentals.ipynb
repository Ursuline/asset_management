import os
from path import Path

cache_dir = './cache/'
data_type = 'stock_data_detailed'
suffix = 'pkl'

def build_cache_path(directory, basename, suffix):
    '''Builds full path from directory, base filename & suffix'''
    os.makedirs(cache_dir, exist_ok=True)
    return os.path.join(cache_dir, f'{basename}.{suffix}')
 

def build_cache_basename(ticker, datatype, expiration_date):
    '''Builds cache base filename of the form: ticker_datatype_expiration'''
    return f'{ticker}_{datatype}_{expiration_date}'


def extract_cache_expiration(path:str):
    '''
    Returns expiration date from file path
    expects base filename of the form ticker_datatype_expiration
    '''
    date_format = "%Y-%m-%d"
    try:
        # the expiration date is the last element in the base filename
        return dt.datetime.strptime(Path(path).stem.split('_')[-1], 
                                    date_format)
    except ValueError:
        print('*** Cannot extract date from filename ***')
        return -1

path = build_cache_path(cache_dir, 
                        build_cache_basename(ticker, data_type, expiration_date), 
                        suffix)

print(f'path= {path}')
dt_exp = extract_cache_expiration(path)
print(f'expiration: {dt_exp}')

```python
start_year = 2010
end_year   = 2020

years      = np.arange(start_year, end_year+1, step=1, dtype=int)
years      = [str(y) for y in years] # convert to string
```

```python
#Balance sheet
asset_map     = [cie.get_balance_sheet_item(item='totalAssets', year=year, change=False) for year in years]
liability_map = [cie.get_balance_sheet_item(item='totalLiabilities', year=year, change=False) for year in years]
equity_map    = [cie.get_balance_sheet_item(item='totalStockholdersEquity', year=year, change=False) for year in years]
leverage_map  = [cie.get_leverage(year=year, change=False) for year in years]

#Income statement
revenue_map        = [cie.get_income_statement_item('revenue', year, change=False) for year in years]
revenue_growth_map = [cie.get_income_statement_item('revenue', year, change=True) for year in years]
ebitda_map         = [cie.get_income_statement_item('ebitda', year, change=True) for year in years]
ebit_map = [cie.get_ebit(year, change=False) for year in years]

# Cash flow statement
fcf_map = [cie.get_cash_flow_statement_item('freeCashFlow', year, change=False) for year in years]

# Financial ratios
roe_map            = [cie.get_financial_ratios_item('returnOnEquity', year, change=False) for year in years]
roa_map            = [cie.get_financial_ratios_item('returnOnAssets', year, change=False) for year in years]
p2b_map            = [cie.get_financial_ratios_item('priceToBookRatio', year, change=False) for year in years]
div_yield_map      = [cie.get_financial_ratios_item('dividendYield', year, change=False) for year in years]
asset_turnover_map = [cie.get_financial_ratios_item('assetTurnover', year, change=False) for year in years]

# Key metrics
per_map           = [cie.get_key_metrics_item('peRatio', year, change=False) for year in years]
debt_2_eq_map     = [cie.get_key_metrics_item(item='debtToEquity', year=year, change=False) for year in years]
current_ratio_map = [cie.get_key_metrics_item(item='currentRatio', year=year, change=False) for year in years]

net_profit_map = [a/b for a,b in zip(fcf_map,revenue_map)]
fcf_2_revenue_map = [a/b for a,b in zip(fcf_map,revenue_map)]
ebit_2_revenue_map = [a/b for a,b in zip(ebit_map,revenue_map)]
ebitda_2_revenue_map = [a/b for a,b in zip(ebitda_map,revenue_map)]
```

```python
dict = {#balance sheet
        'assets': asset_map,
        'liabilities': liability_map,
        'equity': equity_map,
        'debt_to_equity': debt_2_eq_map,
        'leverage': leverage_map,
        # income 
        'revenue': revenue_map,
        'revenue_growth': revenue_growth_map,
        'ebit': ebit_map,
        'ebit_to_revenue': ebit_2_revenue_map,
        'ebitda': ebitda_map,
        'ebitda_to_revenue': ebitda_2_revenue_map,
        'free_cash_flow': fcf_map,
        'FCF_to_revenue': fcf_2_revenue_map,
        'asset_turnover': asset_turnover_map,
        # profitability
        'roe': roe_map,
        'roa': roa_map,
        'p_e_ratio': per_map,
        'price_to_book': p2b_map,
        # solvency
        'current_ratio': current_ratio_map,
        # Stock
        'dividend_yield': div_yield_map,
        }
df = pd.DataFrame(dict).T
df.columns = years
print(df)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import Legend, ColumnDataSource, NumeralTickFormatter, Range1d, LinearAxis
from bokeh.transform import dodge
from bokeh.io import curdoc
import bokeh.palettes as palette
```

```python
curdoc().theme = 'dark_minimal'

line_width = 2
circle_size = 8
circle_fill_color = 'white'
circle_line_color = 'black'
grid_line_color = 'grey'
band_line_color = 'grey'

def title_config(plot):
    plot.title.text_font_size = '20pt'
    return plot

def legend_config(plot):
    plot.add_layout(Legend(), 'right')
    plot.legend.label_text_font_size = '8pt'
    plot.legend.label_width = 75
    plot.legend.border_line_alpha = 0.
    return plot

def grid_config(plot):
    plot.ygrid.minor_grid_line_alpha = .25
    plot.ygrid.minor_grid_line_color = grid_line_color
    plot.ygrid.band_fill_alpha = 0.25
    plot.ygrid.band_fill_color = band_line_color
    return plot

def axes_config(plot, fmt):
    plot.xaxis.axis_label_text_font_size = "10pt"
    plot.yaxis.major_label_orientation = "vertical"
    plot.yaxis[0].formatter = NumeralTickFormatter(format=fmt)
    plot.yaxis.axis_label_text_font_size = "10pt"
    return plot
```

```python
def plot_revenue_ratios():
    title = f'{name} ({company})'
    y_axis_label = 'proportion of revenue'

    plot = figure(height=750, width=1000,
                  x_range=years, 
                  y_range=(0, 1),
                  title=title, 
                  x_axis_label='',
                  y_axis_label=y_axis_label,
                  toolbar_location="above", 
                  tools="box_zoom, pan,reset,save", 
                  active_drag="box_zoom"
                 )
    plot = title_config(plot)
    plot = legend_config(plot)
    plot = axes_config(plot, fmt='0%')
    plot = grid_config(plot)

    y = df.loc['ebitda_to_revenue']
    plot.line(x=years, y=y, legend_label='EBITDA/revenue', width=line_width, color=palette.Set2[8][1])
    plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)

    y = df.loc['ebit_to_revenue']
    plot.line(x=years, y=y, legend_label='EBIT/revenue', width=line_width, color=palette.Set2[8][2])
    plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)

    y = df.loc['FCF_to_revenue']
    plot.line(x=years, y=y, legend_label='FCF/revenue', width=line_width, color=palette.Set2[8][7])
    plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)
    show(plot)

#y = df.loc['asset_turnover']
#plot.line(x=years, y=y, legend_label='Asset turnover', width=line_width, color=palette.Set2[8][5])
#plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)
```

```python
def plot_price_ratios():
    title = f'{name} ({company})'
    y_axis_label = 'ratio'

    plot = figure(height=750, width=1000,
                  x_range=years,
                  title=title, 
                  x_axis_label='',
                  y_axis_label=y_axis_label,
                  toolbar_location="above", 
                  tools="box_zoom, pan,reset,save", 
                  active_drag="box_zoom"
                 )
    plot = title_config(plot)
    plot = legend_config(plot)
    plot = axes_config(plot, '0f')
    plot = grid_config(plot)

    y = df.loc['p_e_ratio']
    plot.line(x=years, y=y, legend_label='P/E', width=line_width, color=palette.Set2[8][1])
    plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)

    y = df.loc['price_to_book']
    plot.line(x=years, y=y, legend_label='price to book', width=line_width, color=palette.Set2[8][2])
    plot.circle(x=years, y=y, line_color=circle_line_color, fill_color=circle_fill_color, size=circle_size)

    show(plot)
```

```python
def plot_revenue():
    title = f'{name} ({company})'
    y_axis_label = f'revenue (billion {currency_symbol})'

    plot = figure(height=750, width=1000,
                  x_range=years, 
                  title=title, 
                  x_axis_label='',
                  y_axis_label=y_axis_label,
                  toolbar_location="above", 
                  tools="box_zoom, pan,reset,save", 
                  active_drag="box_zoom"
                 )
    plot = title_config(plot)
    plot = axes_config(plot, fmt='0g')
    plot = grid_config(plot)

    y = df.loc['revenue']/1e9
    y_color = palette.Set2[8][1]
    plot.line(x=years, y=y, legend_label='revenue', width=line_width, color=y_color)
    plot.circle(x=years, y=y, line_color=y_color, fill_color=circle_fill_color, size=circle_size)
    
    
    y = df.loc['assets']/1e9
    y_color = palette.Set2[8][2]
    plot.line(x=years, y=y, legend_label='assets', width=line_width, color=y_color)
    plot.circle(x=years, y=y, line_color=y_color, fill_color=circle_fill_color, size=circle_size)
    
    
    y2_color = palette.Set2[8][7]
    y2 = df.loc['revenue_growth']
    plot.extra_y_ranges = {"y2": Range1d(start = 0, end = 1)}
    plot.add_layout(LinearAxis(y_range_name = "y2", axis_label='growth'), 'right')
    plot.line(x=years, y=y2, color=y2_color, y_range_name = "y2", legend_label='revenue growth')
    plot.circle(x=years, y=y2, fill_color=circle_fill_color, color=y2_color, y_range_name="y2", size=circle_size)
    
    y3 = df.loc['asset_turnover']
    y3_color = palette.Set2[8][6]
    plot.line(x=years, y=y3, color=y3_color, y_range_name = "y2", legend_label='asset turnover')
    plot.circle(x=years, y=y3, fill_color=circle_fill_color, color=y3_color, y_range_name="y2", size=circle_size)
    
    
    
    plot.yaxis[1].formatter = NumeralTickFormatter(format="2%")
    
    
    
    plot.legend.location = 'top_left'

    show(plot)
```
